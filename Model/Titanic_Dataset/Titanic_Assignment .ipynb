```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer


class AgeImputer(BaseEstimator, TransformerMixin):

    def __init__(self):
        self.imputer = SimpleImputer(strategy='mean') # Initialize the imputer here

    def fit(self, X, y=None):
      # Fit the imputer to the 'Age' column in the training data
      self.imputer.fit(X[['Age']])
      return self # Return self to adhere to scikit-learn's fit method signature

    def transform(self , X):
      # Transform the 'Age' column using the fitted imputer
      X['Age'] = self.imputer.transform(X[['Age']])
      return X
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.base import BaseEstimator, TransformerMixin

class FeatureEncdoer(BaseEstimator , TransformerMixin):

  def fit(self  , X , y=None):
    return self

  def transform(self , X):
    # Instantiate OneHotEncoder before using it
    encoder = OneHotEncoder()
    matrix = encoder.fit_transform(X[['Embarked']]).toarray()

    # Add checks for the actual columns created by the encoder
    # and handle potential missing categories if necessary.
    # For simplicity here, we assume the expected categories are present.
    columns_name = ['C', 'Q', 'S','N']
    # Ensure the number of columns matches the encoded matrix
    if len(columns_name) != matrix.shape[1]:
        # Handle unexpected number of encoded columns, e.g., print a warning
        print(f"Warning: Expected {len(columns_name)} columns for Embarked, got {matrix.shape[1]}. Adjusting column names.")
        columns_name = [f'Embarked_{i}' for i in range(matrix.shape[1])]


    for i in range(len(matrix.T)):
      X[columns_name[i]] = matrix.T[i]

    # Instantiate a new OneHotEncoder for the 'Sex' column
    encoder_sex = OneHotEncoder()
    matrix = encoder_sex.fit_transform(X[['Sex']]).toarray()
    columns_name = ['Female', 'Male']

    # Ensure the number of columns matches the encoded matrix
    if len(columns_name) != matrix.shape[1]:
        # Handle unexpected number of encoded columns, e.g., print a warning
        print(f"Warning: Expected {len(columns_name)} columns for Sex, got {matrix.shape[1]}. Adjusting column names.")
        columns_name = [f'Sex_{i}' for i in range(matrix.shape[1])]


    for i in range(len(matrix.T)):
      X[columns_name[i]] = matrix.T[i]

    return X
```

```python
class FeatueDroper():

  def fit(self, X , y=None):
    return self

  def transform(self , X):
    return X.drop(['Embarked', 'Name' , 'Ticket', 'Sex', 'Cabin', 'N'], axis=1,errors='ignore')
```

```python
from sklearn.pipeline import  Pipeline

pipeline = Pipeline([
    ('ageimputer' , AgeImputer()),
    ('encoder' , FeatureEncdoer()),
    ('droper' , FeatueDroper())
])
```

```python
X_train = pipeline.fit_transform(X_train)
```
