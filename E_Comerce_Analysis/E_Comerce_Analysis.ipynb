```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
import seaborn as sns
```

```python
raw_data = pd.read_csv('E_Comerce.csv' , encoding='latin1')
raw_data.head(2)
```

```python
raw_data.describe()
```

```python
raw_data.info()
```

#Converting Date Time

```python
raw_data['Order Date'] = pd.to_datetime(raw_data['Order Date'])
raw_data['Ship Date'] = pd.to_datetime(raw_data['Ship Date'])
```

```python
raw_data.info()
```

```python
raw_data.head(2)
```

```python
raw_data['Order_Month'] = raw_data['Order Date'].dt.month
raw_data['Order_Year'] = raw_data['Order Date'].dt.year
raw_data['Order_Day_Of_Week'] = raw_data['Order Date'].dt.day_of_week
```

```python
raw_data.head(2)
```

#Monthly Sales Analysis

```python
sales_by_month = raw_data.groupby('Order_Month')['Sales'].sum().reset_index()

sales_by_month
```

```python
fig = px.line(sales_by_month,
              x='Order_Month',
              y='Sales',
              title='Monthly Sales Analysis')
fig.show()
```

#Sales Analysis by Category

```python
raw_data['Category'].unique()
```

```python
sales_by_category = raw_data.groupby('Category')['Sales'].sum()
# highest_sales = raw_data.groupby('Category')['Sales'].sum().idxmax()
sales_by_category
```

```python
fig = px.pie(
    sales_by_category,
    values='Sales',
    names=sales_by_category.index,
    hole=0.3,
    title='Sales by Category',
)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_font=dict(size=24))

fig.show()
```

#Sales Analysis By Sub_Category

```python
raw_data['Sub-Category'].unique()
```

```python
sales_by_sub_category = raw_data.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).reset_index()
sales_by_sub_category
```

```python
fig = px.bar(
    sales_by_sub_category,
    x='Sub-Category',
    y='Sales',
    title='Sales by Sub-Category')

fig.show()
```

#Monthly Profit

```python
raw_data.head(2)
```

```python
profit_by_month = raw_data.groupby('Order_Month')['Profit'].sum().reset_index()
profit_by_month
```

```python
# prompt: Make a line Graph for profit_by_month

fig = px.line(profit_by_month,
              x='Order_Month',
              y='Profit',
              title='Monthly Profit Analysis')
fig.show()
```

```python
# prompt: Make a Bar Graph for profit by moonth

# Profit by Month Bar Graph
fig = px.bar(profit_by_month,
             x='Order_Month',
             y='Profit',
             title='Profit by Month',
             labels={'Order_Month': 'Month', 'Profit': 'Profit'})
fig.show()
```

#Profit By Category

```python
profi_by_category = raw_data.groupby('Category')['Profit'].sum().sort_values(ascending=False).reset_index()
profi_by_category
```

```python
fig = px.pie(
    profi_by_category,
    values='Profit',
    names='Category',
    hole=0.3,
    title='Profit by Category',
)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_font=dict(size=24))

fig.show()
```

#Profit_By_Sub_Category


```python
profit_by_dub_category = raw_data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).reset_index()
profit_by_dub_category
```

```python
# prompt: Generate a bar graph and show the profit by sub category

# Profit by Sub-Category Bar Graph
fig = px.bar(profit_by_dub_category,
             x='Sub-Category',
             y='Profit',
             title='Profit by Sub-Category',
             labels={'Sub-Category': 'Sub-Category', 'Profit': 'Profit'})
fig.show()
```

#Sales And Profit - With Customer Segment

```python
raw_data.head(2)
```

```python
profit_by_segment = raw_data.groupby('Segment')['Profit'].sum().sort_values(ascending=False).reset_index()
profit_by_segment
```

```python
profit_by_segment = raw_data.groupby('Segment').agg({'Profit': 'sum', 'Sales': 'sum'}).reset_index()
profit_by_segment
```

```python
color_palette = ['#1f77b4', '#ff7f0e']

fig = go.Figure()
fig.add_trace(go.Bar(
             x=profit_by_segment['Segment'],
             y=profit_by_segment['Profit'],
             name='Profit',
             marker_color=color_palette[0]
))

fig.add_trace(go.Bar(
             x=profit_by_segment['Segment'],
             y=profit_by_segment['Sales'],
             name='Sales',
             marker_color=color_palette[1]
))

fig.update_layout(title='Sales And Profit Analysis By Customer Segment')

fig.show()
```

#Sales to Profit Ratio

```python
sales_profit_ratio = raw_data.groupby('Segment').agg({'Profit': 'sum', 'Sales': 'sum'}).reset_index()
sales_profit_ratio['Sales_Profit_Ratio'] = sales_profit_ratio['Sales'] / sales_profit_ratio['Profit']

print(sales_profit_ratio[['Segment','Sales_Profit_Ratio']])
```
